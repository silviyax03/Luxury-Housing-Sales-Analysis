```python
# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
# Check dataset shape before removing duplicates
print("Shape before removing duplicates:", df.shape)

# Remove duplicate rows
df = df.drop_duplicates()

# Check dataset shape after removing duplicates
print("Shape after removing duplicates:", df.shape)
```

```python
# Check duplicate rows
df.duplicated().sum()
```

```python
df["Ticket_Price_Cr"].head(10)
```

```python
# Remove ₹ and Cr, then convert to numeric
df["Ticket_Price_Cr"] = (
    df["Ticket_Price_Cr"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Cr", "", regex=False)
    .str.strip()
)

# Convert the column to numeric
df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors="coerce")
```

```python
# Convert Configuration values to uppercase
df["Configuration"] = df["Configuration"].str.upper()
```

```python
df["Sales_Channel"] = df["Sales_Channel"].str.strip()
```

```python
df["NRI_Buyer"].value_counts()
```

```python
# Standardize Yes/No values
df["NRI_Buyer"] = (
    df["NRI_Buyer"]
    .str.strip()
    .str.title()
)
```

```python
df["NRI_Buyer"].value_counts()
```

```python
df[["Unit_Size_Sqft", "Ticket_Price_Cr", "Amenity_Score"]].describe()
```

```python
# Replace invalid values with NaN
df.loc[df["Unit_Size_Sqft"] <= 0, "Unit_Size_Sqft"] = np.nan
df.loc[df["Ticket_Price_Cr"] <= 0, "Ticket_Price_Cr"] = np.nan
```

```python
df[["Unit_Size_Sqft", "Ticket_Price_Cr"]].describe()
```

```python
# Fill missing values with median
df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].median(), inplace=True)

df["Ticket_Price_Cr"].fillna(df["Ticket_Price_Cr"].median(), inplace=True)

df["Amenity_Score"].fillna(df["Amenity_Score"].median(), inplace=True)
```

```python
df["Buyer_Comments"].fillna("No Comments", inplace=True)
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Price per Sqft (in Crores per Sqft)
df["Price_per_Sqft"] = df["Ticket_Price_Cr"] / df["Unit_Size_Sqft"]
```

```python
df[["Ticket_Price_Cr","Unit_Size_Sqft","Price_per_Sqft"]].head()
```

```python
df["Purchase_Quarter"] = pd.to_datetime(df["Purchase_Quarter"])
```

```python
df["Quarter_Number"] = df["Purchase_Quarter"].dt.quarter
```

```python
df[["Purchase_Quarter","Quarter_Number"]].head()
```

```python
df["Purchase_Year"] = df["Purchase_Quarter"].dt.year
df["Quarter_Label"] = "Q" + df["Quarter_Number"].astype(str)
```

```python
df[["Purchase_Quarter","Quarter_Number","Purchase_Year","Quarter_Label"]].head()
```

```python
df.to_csv("Luxury_Housing_Cleaned.csv", index=False)

print("Cleaned dataset saved successfully!")
```
